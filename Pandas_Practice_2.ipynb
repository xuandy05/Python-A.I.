```
import pandas as pd
pd.set_option('display.expand_frame_repr', False)

url = "https://pynative.com/wp-content/uploads/2019/01/Automobile_data.csv"
```

1. Using the url provided use the read_csv function to create a data frame.
print the first 5 elements using the head() function, and the last 5 elements using the tail function


```
df = pd.read_csv(url)
print(df.head())
print(df.tail())
```

2. Now use boolean indexing to find the longest car in the set, then find the cheapest car in the set.

```
longest = max(df['length']) 
longest_car = df[df['length'] == longest]
print(longest_car)
cheapest = min(df['price'])
cheapest_car = df[df['price'] == cheapest]
print(cheapest_car)
```

3. Print the number of cars for each company

```
x = df[df['company'] == 'alfa-romero']
x.size/10
```

4. Add a new column to the end of the Dataframe. The name of this column is 'horse price' and the value of this column is equal to the price divided its horse power. (price/horsepower)

```
df.insert(10, "horse price", df['price']/df['horsepower'], True)
df
```

### BONUS
5. Delete any rows that contains the value NaN

```
df = df.dropna()
df.reset_index(drop = True)
```

6. Repalce all NaN values with the value 1

```
import numpy as np
df = df.replace(np.nan, 1)
df
# df = fillna(1)
```

```
describe = df.describe()
features = ['wheel-base', 'length', 'horsepower', 'average-mileage', 'price', 'horse price']
for each in features:
  IQR = describe[each]['75%'] - describe[each]['25%']
  df = df[df[each] < describe[each]['75%'] + 1.5 * IQR]
  df = df[df[each] > describe[each]['25%'] - 1.5 * IQR]
df = df.reset_index(drop = True)
df
```
